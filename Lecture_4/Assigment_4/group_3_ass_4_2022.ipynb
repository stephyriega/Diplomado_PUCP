3. What are the top 5 banks each year with the highest interest rate at `Préstamos hipotecarios para vivienda`, `Consumo -

```python
#We replace the mispelled column names in "TASA ANUAL (%)" and homogenize the values
data_bank["TASA ANUAL (%)"] = data_bank["TASA ANUAL (%)"]. replace([ "Tarjetas de crédito", "Tarjetas de crédito abiertas4",  "Tarjetas de crédito cerradas4",
                                                          "Tarjetas de crédito cerradas5", "Tarjetas de crédito cerradas", "Tarjetas de crédito abiertas3" ], "Tarjetas de Crédito")
data_bank["TASA ANUAL (%)"] = data_bank["TASA ANUAL (%)"]. replace([ "Hipotecarios", "HIPOTECARIO", 
                                                          "PRÉSTAMOS HIPOTECARIOS" ], "Préstamos hipotecarios para vivienda")
#Now we create new variables, change the format wide and create a new dataframe with the top 5 for year and concept
ranking = data_bank.copy().loc[data_bank["TASA ANUAL (%)"].isin(["Préstamos hipotecarios para vivienda","Tarjetas de Crédito"])]
ranking.drop(["DATE_INFO", "PROMEDIO"], axis = "columns", inplace = True)
ranking = ranking.melt(id_vars = ["YEAR", "TASA ANUAL (%)"])
ranking.columns = ["YEAR", "rate_concept", "banks", "rate_value"]
ranking = ranking.sort_values("rate_value", ascending = False).groupby(["YEAR","rate_concept"]).head(5).reset_index(drop = True).sort_values(["YEAR","rate_concept","rate_value"], ascending = [True, True, False])

ranking
```
