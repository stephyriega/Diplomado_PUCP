## Second method

```python
data_vacunacion = data[ data.iloc[ :, 3 ].notnull() ].copy()
data_vacunacion
```

```python
def geo_vacunacion( row_series ):
    
    address = f"{row_series['Departamento']}, {row_series['Provincia']}, {row_series['Distrito']}, {row_series['Direccion']}"
    
    # Set Geolocation
    gmaps = googlemaps.Client( API_KEY )
    result_api = gmaps.geocode( address , region = 'PE' )
    
    # Information
    try:
        lat = result_api[0]['geometry']['location']['lat']
        lon = result_api[0]['geometry']['location']['lng']  
    except:
        lat = np.nan
        lon = np.nan
    
    return ( address, lat, lon )
```

```python
data_vacunacion['results'] = data_vacunacion.apply( lambda x: geo_vacunacion( x )  , axis = 1 )
data_vacunacion
```

```python
data_vacunacion[['Address', 'Latitude', 'Longitude']] = pd.DataFrame( data_vacunacion.results.tolist() ,  index = data_vacunacion.index )
```

```python
output = data_vacunacion [['Address','Latitude','Longitude']]
output
```
