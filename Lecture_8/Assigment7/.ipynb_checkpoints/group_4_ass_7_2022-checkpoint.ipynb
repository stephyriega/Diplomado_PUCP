### 1.2 Importar datos de PDF

```python
dbanco = tabula.read_pdf("https://www.bbva.pe/content/dam/public-web/peru/documents/personas/canales-de-atencion/oficinas/Oficinas-BBVA-abiertas-23.10.20.pdf", pages="all", stream=True)
```

```python
# Observamos que hay un error al importar la página 2
print([len(i.columns) for i in dbanco])
```

```python
print(len(dbanco))
dbanco[0]
```

```python
# Se elimina columna vacía
dbanco[1].dropna(how='all', axis=1, inplace=True)
```

```python
for i in range(len(dbanco)):
    dbanco[i] = dbanco[i].reset_index().T.reset_index().T.reset_index()
    dbanco[i].drop(dbanco[i].iloc[:, 0:2], axis = 1, inplace=True)
    dbanco[i].set_axis(["oficina", "direccion", "distrito", "provincia", "departamento"], axis=1, inplace=True)
```

```python
# En la data de la primera hoja se elimina las primeras dos columnas mal importadas
dbanco[0]= dbanco[0].iloc[2:, :]
```

```python
# Se concatena toda la data
bbva_peru = pd.concat(dbanco, ignore_index= True)
```

```python
# Nos quedamos solo con Lima
bbva_peru["departamento"].replace(["LIMA.1"], ["LIMA"], inplace =   True)
print(bbva_peru['departamento'].value_counts())
bbva_lima = bbva_peru.loc[bbva_peru["departamento"] == "LIMA", :]
```

```python
bbva_lima
```

```python
cols = ['direccion', 'distrito']
bbva_lima[ 'complete_address' ] = bbva_lima[ cols ].apply(lambda row: ', '.\
                                  join( row.values.astype( str ) ) , axis = 1 )
```

```python
bbva_lima
```

# Parte 2. Geocode

```python
# Importamos el paquete correspondiente y seteamos el API
import googlemaps
gmaps = googlemaps.Client( key = 'AIzaSyCOeOz6pzPmjHcGmQkFXvuwUm1_j9PcWgw' )
```

```python
# Definimos la función para la geolocalización

def get_results( result_api ):
    
    try:
        lat = result_api[0]['geometry']['location']['lat']
        lon = result_api[0]['geometry']['location']['lng']  
        
    except:
        lat = np.nan
        lon = np.nan
    
    return ( lat, lon )
```

```python
# Aplicaciones la función anteriormente definina mediante el método lambda. 
# Se crea una nueva columna denominada "results", que alberga la latitud y longitud de las oficinas

bbva_lima[ 'results' ] = bbva_lima.apply( lambda x: get_results( gmaps.geocode( x[ 'complete_address' ] , region = 'pe' ) ) , axis = 1 )
```

```python
# Verificamos el porcentage de oficinas que no pudieron ser geolocalizadas

bbva_lima[ 'results' ].value_counts( normalize = True )
```

```python
# Desdoblamos la columna "results" en columnas separadas de latitud y longitud

bbva_lima[ [ 'latitude' , 'longitud' ] ] = pd.DataFrame( bbva_lima.results.tolist(), index = bbva_lima.index )
```

```python
# Identificamos las filas en las que las oficinas no pudieron ser geolocalizadas

bbva_lima[ bbva_lima [ 'latitude' ].isna() ]
```
