### Answer:

We create a function that imports the files that belong to a certain range of days. It use four arguments, two of them are required. The function take the arguments and define a start day and a end day. First, it search for all the `.xlsx` files in the directory. Then, take a fraction of each file's name to convert it into a date associated to the file. Given the start day and end day selected, the function chooses which dates belongs to the range of days selected. Finally, the files associated to those dates are uploaded. <br>
<br>
If no `end_month` neither `end_year` is given, the function takes automatically the present month and year as limit of the range. <br>
<br>
The function use functions and methods from the `os`, `re`, `datetime`, `calendar` and `pandas` libraries.

```python
import os
import re
import datetime
import calendar
import pandas as pd
```

```python
def import_data_sbs_banks(start_month, start_year, end_month = datetime.date.today().month, end_year = datetime.date.today().year):
    path = '../../_data/sbs/B_RawData/bancos'
    # List of xlsx files in the folder
    files = [f for f in os.listdir(path) if re.search("\\.xlsx$",f)]
    # List of dates corresponding to each xlsx file
    dates = [*map(lambda x: datetime.datetime.strptime(x[12:-5], "%d_%m_%Y").date(), files)]
    # Exact date to start importing
    start = datetime.datetime.strptime("01-"+str(start_month)+"-"+str(start_year), "%d-%m-%Y").date()
    # Define the last day of the month
    end_day = calendar.monthrange(int(end_year), int(end_month))[1]
    # Exact date to stop importing
    end = datetime.datetime.strptime(str(end_day)+"-"+str(end_month)+"-"+str(end_year), "%d-%m-%Y").date()
    # List of files that satisfy the date condition
    files = [files[i] for i in range(len(dates)) if start <= dates[i] <= end]
    # Dictionary with the selected files imported
    final = dict(map(lambda x: (x[12:-5], pd.read_excel(path+"/"+x)), files))
    return final
```
