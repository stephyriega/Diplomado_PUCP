2. Generate ubigeo for Departments and Provinces taking the first two and four numbers. Hint: Use this code.

6. Use geopandas to plot the number of cases by the department for all 2021 quarters using subplots. Every subplot for each quarter. Use a categorical legend with 5 bins. Do not forget to indicate the color of NA values. Use this shapefile. For this task, you will have to aggregate shapefiles at the department level. Hint: Use Semana variable to group by quarters.

```python
Dengue_semanal = Dengue_peru.loc[:,["Año", "Semana", "Casos", "Ubigeo"]]

Dengue_semanal["Trimestre"] = pd.Series(dtype=int)

for index,row in Dengue_semanal.iterrows():
    if Dengue_semanal.loc[index,"Semana"] <= 13:
        Dengue_semanal.loc[index,"Trimestre"] = 1
    elif Dengue_semanal.loc[index,"Semana"] <=26:
        Dengue_semanal.loc[index,"Trimestre"] = 2
    elif Dengue_semanal.loc[index,"Semana"] <=39:
        Dengue_semanal.loc[index,"Trimestre"] = 3
    else:
        Dengue_semanal.loc[index,"Trimestre"] = 4

for trim in range(1,5):

    alterna = Dengue_semanal[Dengue_semanal["Trimestre"]==trim]

    distritos_merge=pd.merge(distritos, alterna, how="left", on=['Ubigeo'])

    dpto_shp = distritos_merge.dissolve( by = 'CCDD' )

    ax = plt.subplot(2, 2, trim)

    dpto_shp.plot(column='Casos',
                  cmap='Reds',
                  figsize=(20, 20),
                  legend = True, linestyle='-',
                  edgecolor='black',
                  missing_kwds= dict(color = "#72a0c1"),
                  ax=ax)

    ax.set_title(f'2021-T{trim}')
    ax.set_axis_off()
plt.show()
```
