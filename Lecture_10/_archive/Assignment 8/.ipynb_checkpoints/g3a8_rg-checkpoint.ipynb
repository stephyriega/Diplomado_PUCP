# 1. Create JN (and import libraries)

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import geopandas as gpd
```

# 2. Import Data

```python
dengue = pd.read_csv( r'../../_data/data_dengue_peru.csv', thousands = ',', dtype = { 'Ubigeo': 'str', 'Año': 'str' } )
```

# 3. Generate ubigeo for departments and provinces

```python
dengue[ 'dep_id' ] = dengue.Ubigeo.str[ : 2 ].copy()
dengue[ 'prov_id' ] = dengue.Ubigeo.str[ : 4 ].copy()
```

# 4. Plot cases per district in 2021

```python
dist_shp = gpd.read_file( r'../../_data/LIMITE_DISTRITAL_2020_INEI/INEI_LIMITE_DISTRITAL.shp' )
```

```python
dengue_2021 = dengue[ dengue[ 'Año' ] == '2021' ]
```

```python
dengue_2021_1 = dengue_2021.groupby( [ 'Ubigeo' ], as_index = False )[ [ 'Casos' ] ].sum()
```

```python
dg_map1 = dist_shp.merge( dengue_2021_1, left_on = 'UBIGEO', right_on = 'Ubigeo', how = 'left' )
```

```python
fig, ax = plt.subplots( figsize = ( 20, 20 ) )

dg_map1.plot( ax = ax, column = 'Casos', cmap = 'Reds', linestyle = 'dotted', edgecolor = 'black', legend = True, 
             missing_kwds = dict( color = '#D0D0D0' ) )
```

# 5. Plot cases per province in 2021

```python
prov_shp = dist_shp.dissolve( by = 'IDPROV', as_index = False )
```

```python
dengue_2021_2 = dengue_2021.groupby( [ 'prov_id' ], as_index = False )[ [ 'Casos' ] ].sum()
```

```python
dg_map2 = prov_shp.merge( dengue_2021_2, left_on = 'IDPROV', right_on = 'prov_id', how = 'left' )
```

```python
fig, ax = plt.subplots( figsize = ( 20, 20 ) )

dg_map2.plot( ax = ax, column = 'Casos', cmap = 'Reds', linestyle = 'dotted', edgecolor = 'black', legend = True, 
             missing_kwds = dict( color = '#D0D0D0' ) )
```

# 6. Plot cases per department for all years

```python
dep_shp = dist_shp.dissolve( by = 'CCDD', as_index = False )
```

```python
dengue1 = dengue.groupby( [ 'Año', 'dep_id' ], as_index = False)[ [ 'Casos' ] ].sum()
```

```python
dg_map3 = dep_shp.merge( dengue1, left_on = 'CCDD', right_on = 'dep_id', how = 'left' )
```

```python
fig, axis = plt.subplots( nrows = 3, ncols = 3, figsize = ( 20, 20 ) )

cmap = plt.cm.OrRd

idx = 0

for i in range( 3 ):
    if i < 2:
        for j in range( 3 ):
            ax = axis[ i ][ j ]
            year = dg_map3.Año.unique()[ idx ]
            dg_temp = dg_map3[ ( ( dg_map3[ 'Año' ] == year ) | ( dg_map3[ 'Año' ].isna() ) ) ]
            dg_temp.plot( column = 'Casos', cmap = cmap, linestyle = '--', edgecolor = 'black', ax = ax, legend = True, missing_kwds = dict( color = '#D0D0D0' ) )
            ax.set_title( year )
            idx = idx + 1
    else:
        ax = axis[ i ][ 1 ]
        year = dg_map3.Año.unique()[ idx ]
        dg_temp = dg_map3[ ( ( dg_map3[ 'Año' ] == year ) | ( dg_map3[ 'Año' ].isna() ) ) ]
        dg_temp.plot( column = 'Casos', cmap = cmap, linestyle = '--', edgecolor = 'black', ax = ax, legend = True, missing_kwds = dict( color = '#D0D0D0' ) )
        ax.set_title( year )
        idx = idx + 1
```

```python
def Q_id( n_week ):
    if n_week < 14:
        quarter = 1
    elif 14 <= n_week < 27:
        quarter = 2
    elif 27 <= n_week < 40:
        quarter = 3
    elif n_week >= 40:
        quarter = 4
    else: 
        quarter = float( "nan" )
    return quarter
```

```python
dengue_2021.[ 'Q' ] = dengue_2021.apply( lambda row: Q_id( row[ 'Semana' ] ), axis = 1 )
```

```python
dengue_Q = dengue_2021.groupby( [ 'Q', 'dep_id' ], as_index = False )[ [ 'Casos' ] ].sum()
```

```python
dg_map4 = dep_shp.merge( dengue_Q, left_on = 'CCDD', right_on = 'dep_id', how = 'left', validate = '1:m' )
```

```python
bins = pd.qcut( dg_map4[ 'Casos' ], 5, retbins = True )[ 1 ][ 1: ]
```

```python
fig, axis = plt.subplots( nrows = 2, ncols = 2, figsize = ( 20, 20 ) )

idx = 0

for i in range( 2 ):
    for j in range( 2 ):
        ax = axis[ i ][ j ]
        quarter = dg_map4.Q.unique()[ idx ]
        dg_temp = dg_map4[ ( ( dg_map4[ 'Q' ] == quarter ) | ( dg_map4[ 'Q' ].isna() ) ) ]
        dg_temp.plot( column = 'Casos', cmap = cmap, linestyle = '--', edgecolor = 'black', ax = ax, legend = True, 
                      scheme = 'User_defined',
                      missing_kwds = dict( color = '#D0D0D0' ), 
                      classification_kwds = dict( bins = bins ), 
                      legend_kwds = dict( loc = 'upper left', bbox_to_anchor = ( 1.01, 1 ), title = 'Number of Cases', 
                                          frameon = False, fontsize = 'x-large', title_fontsize = 'x-large' ) )
        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticks([])
        ax.set_title( f"Q{ str( quarter )[ 0 ] }" )
        idx = idx + 1
fig.suptitle( 'Numbers of Cases per Quarter, by Region', fontsize = 20 )
fig.tight_layout()
```
