7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
# Se usa loc para seleccionar las variables solicitadas en el ejemplo, luego se añada el método gruopby para solicitar cualquier output 
# en función a la variable departamente (V024), luego se agrega la media. 
endes_2019_r = endes_2019.loc[ : , ["V201","V613","V715","V511","V024"]].groupby(['V024']).mean()
# Se genera el objeto mean_keys_vars y renombrando la etiqueta de las columnas a través de .rename
mean_key_vars = endes_2019_r.rename(columns={'V201': 'mean_total_children',
                             'V613': 'mean_ideal_children',
                             'V715': 'mean_hb_yr_educ',
                             'V511': 'mean_first_marriage'})
mean_key_vars
```
