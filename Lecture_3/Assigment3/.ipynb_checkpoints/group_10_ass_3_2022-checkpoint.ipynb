7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

#### Solution 7

```python
## Replace the categorical response to 0 in column V613
endes_2019.V613 = endes_2019.V613.replace("Respuesta no numérica", 0)
## Change from categorical column to numerical column. Code obbtained from https://www.statology.org/convert-categorical-variable-to-numeric-pandas/ 
endes_2019['V613'] = pd.factorize(endes_2019['V613'])[0]

## Replace the categorical response to 0 in column V715
endes_2019.V715 = endes_2019.V715.replace("No sabe", 0)
## Change from categorical column to numerical column. Code obbtained from https://www.statology.org/convert-categorical-variable-to-numeric-pandas/ 
endes_2019['V715'] = pd.factorize(endes_2019['V715'])[0]

## Group by year and department, then use aggregate function to get the mean of each column
mean_key_vars = endes_2019.groupby(["year", "V024"]).agg({"V201": "mean", "V613":"mean", "V715" :"mean", "V511": "mean"})
## Change column names
mean_key_vars.columns = ['mean_total_children', 'mean_ideal_children', "mean_hb_yr_educ", "mean_first_marriage"]
## Print
mean_key_vars
```

8. Reshape `mean_key_vars` from wide to long. We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

#### Solution 8

```python
## Reset index to have the "dpto" as a new column
mean_key_vars = mean_key_vars.reset_index(inplace=False)

## we use melt to change the shape from wide to long
## select ID variable the "dpto" column, which is V024 in our case. 
## select value variables as the four columns indicated
reshape_mean_key_vars = mean_key_vars.melt(id_vars = "V024", value_vars = ["mean_total_children", "mean_ideal_children", "mean_hb_yr_educ", "mean_first_marriage"])

## Change the column names as indicated in the exercise
reshape_mean_key_vars.columns = ['dpto', 'variables', "values"]

## Print the new variable
reshape_mean_key_vars
```

9. Replicate your answers for questions 7 and 8, but in one line of code. Make it the most simple as possible. **NO HINT**

#### Solution 9

```python
## Question 7
## V201": "mean", "V613":"mean", "V715" :"mean", "V511"

## Since the categorical variables were converted to numerical in previous exercise, we only need to groupby and get mean of indicated columns
## In a simple unique line, we first group by year and dpto, then calculate mean of all columns.
## Finally, we selected the expected columns and change their names using rename function.
endes_2019.groupby(["year", "V024"]).mean()[["V201", "V613", "V715", "V511"]].rename(columns= {"V201": "mean_total_children", "V613":"mean_ideal_children", "V715": "mean_hb_yr_educ", "V511":"mean_first_marriage"})


## Question 8
## In one line, we used the methods melt, indicating the identifier variable as dpto (V024) and value variables (4 columns)
## Then we changed the name of the three columns using the function rename
mean_key_vars.melt(id_vars = "V024", value_vars = ["mean_total_children", "mean_ideal_children", "mean_hb_yr_educ", "mean_first_marriage"]).rename(columns = {"V024": "dpto", "variable": "variables", "value":"values"})

```

10. Merge `reshape_mean_key_vars` with `endes_2019`. Name this object `final_result`. **Hint: Use merge.**

#### Solution 10

```python
print(reshape_mean_key_vars)
print(endes_2019)

## Use the method merge to merge the two dataframes. 
## Indicate that the Identifier column in endes_2019 is V024, while in reshape_mean_key_vars is dpto.
final_result = pd.merge(endes_2019, reshape_mean_key_vars, left_on = "V024" ,right_on="dpto")

## Print the new database
print(final_result)
```
