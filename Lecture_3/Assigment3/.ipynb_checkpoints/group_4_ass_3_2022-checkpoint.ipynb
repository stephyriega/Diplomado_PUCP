#### Answer 1:
The data is available in the path `"../../_data/endes/2019"`, so we just need to use the `sav` library in order to import all the dataframes. For each dataframe, we use the `pd.reader_spss` because the data is in the format spss. Then, we use the `header.all().varLabels` function to label the variables, and the `header.all().valueLabels` to label the values.

```python
import savReaderWriter as sav
```

```python
rec1 = pd.read_spss( fr"../../_data/endes/2019/REC0111.sav" )
with sav.SavHeaderReader( fr"../../_data/endes/2019/REC0111.sav", ioUtf8=True) as header:
    var_labels1 = header.all().varLabels
    value_labels1 = header.all().valueLabels

rec2 = pd.read_spss( fr"../../_data/endes/2019/RE223132.sav" )
with sav.SavHeaderReader( fr"../../_data/endes/2019/RE223132.sav", ioUtf8=True) as header:
    var_labels2 = header.all().varLabels
    value_labels2 = header.all().valueLabels

rec3 = pd.read_spss( fr"../../_data/endes/2019/RE516171.sav" )
with sav.SavHeaderReader( fr"../../_data/endes/2019/RE516171.sav", ioUtf8=True) as header:
    var_labels3 = header.all().varLabels
    value_labels3 = header.all().valueLabels
```

#### Answer 2:
First, we create three lists containing the variables we want to keep. Then, we create new dataframes using the `loc` method to filter the columns we want to keep and all the rows of the dataset. <br>
To create the new dictionaries for variable labels we use a `for loop` that keeps all the keys that were in `var_labels` dictionary and in the `list` initially created. <br>
To create the new dictionaries for value labels we use two `for loop`s. The inner loop create a list with the names of variables that are in the `values_labels` dictionary and in the list initially created. The outter loop create a dictionary that keeps the keys corresponding to the list we just creat. <br>
This process is repeated for each group of data that we had imported.

```python
list1 = ["CASEID", "V000", "V001", "V002", "V003", "V004", "V007", "V008", "V009", "V010", "V011", "V012", "V024", "V102", "V120", "V121", "V122", "V123", "V124", "V125", "V127", "V133"]
list2 = ["CASEID", "V201", "V218", "V301", "V302", "V323", "V323A", "V325A", "V326", "V327", "V337", "V359", "V360", "V361", "V362", "V363", "V364", "V367", "V372", "V372A", "V375A", "V376", "V376A", "V379", "V380"]
list3 = ["CASEID", "V501", "V502", "V503", "V504", "V505", "V506", "V507", "V508", "V509", "V510", "V511", "V512", "V513", "V525", "V613", "V714", "V715"]
```

```python
rec1_1 = rec1.loc[:,list1]
new_var_labels1 = {key: var_labels1[key] for key in list1}
new_value_labels1 = {key: value_labels1[key] for key in [x for x in list(value_labels1.keys()) if x in list1]}
```

```python
rec2_1 = rec2.loc[:,list2]
new_var_labels2 = {key: var_labels2[key] for key in list2}
new_value_labels2 = {key: value_labels2[key] for key in [x for x in list(value_labels2.keys()) if x in list2]}
```

```python
rec3_1 = rec3.loc[:,list3]
new_var_labels3 = {key: var_labels3[key] for key in list3}
new_value_labels3 = {key: value_labels3[key] for key in [x for x in list(value_labels3.keys()) if x in list3]}
```

#### Answer 3:
To create a new column for the `rec1_1` dataframe, we use the `.loc` function. Then, we use the `.update` function to update the label's dictionary.

```python
rec1_1.loc[:,"year"] = 2019
new_var_labels1.update({"year":"Year of the survey"})
```

#### Answer 4:
We merge the three dataframes using `merge` function using the CASEID column as the key. We use the “outer” method in order to have the union of all cases in the three dataframes

```python
endes_2019 = rec1_1.merge(rec2_1, on="CASEID", how="outer").merge(rec3_1, on="CASEID", how="outer")
endes_2019
```

#### Answer 5:
We create the object var_labels as a copy of the first dictionary (new_var_labels1). We use the update function to insert the second dictionary’s labels into var_labels. Finally, we repeat the same process to insert de third dictionary’s labels. For the value labels, we repeat the same process inserting the three dictionaries’ labels into the object named value_labels.

```python
var_labels = new_var_labels1.copy()
var_labels.update(new_var_labels2)
var_labels.update(new_var_labels3)
var_labels
```

```python
value_labels = new_value_labels1.copy()
value_labels.update(new_value_labels2)
value_labels.update(new_value_labels3)
value_labels
```

```python
endes_2019.attrs["var_labels"] = var_labels
endes_2019.attrs["value_labels"] = value_labels
```

#### Answer 6:
Columns `V613` and `V715` are categorical type and both have `strings` values so is not possible obtain its `min` `max` and `mean` values. Because of that, first we replace the `strings` values with `np.nan`s and convert from categorical type to string type and then to float type. Once this is done, we can continue with the exercise. <br>
We select the four columns desired and use the `describe` method to get the all the information that the function provides. Then, we add a row containing the count of missing values for each variable. Once we have all the statistics we use `loc` method to select only the ones requested and use the `transpose` method to change the rows to columns and the columns to rows. Later, we convert the index (which has all the variable names) into a column using `reset_index`. Finally, we change the names of almost every column to the ones indicated and sort the values by the total numer of missing values.

```python
# Convert V613 and V715 from categorical to float
endes_2019["V613"] = endes_2019["V613"].replace("Respuesta no numérica", np.nan).astype(str).astype(float)
endes_2019["V715"] = endes_2019["V715"].replace("No sabe", np.nan).astype(str).astype(float)
```

```python
table_endes = endes_2019.loc[:,["V201","V613","V715","V511"]].describe(include="all")
table_endes.loc["N_missing"] = endes_2019.loc[:,["V201","V613","V715","V511"]].isna().sum()
table_endes = table_endes.loc[["min","max","mean","count","N_missing"]].transpose().reset_index()
table_endes = table_endes.rename(columns = {"index":"Variables", "min":"Min", "max":"Max","mean":"Mean","count":"N_obs"}).sort_values("N_missing")
table_endes
```

#### Answer 7:
First, we use `groupby` method selecting `V024` as the variable from which we will group the data and indicating that we does not want to convert this column to the `index`. Then, we select the four variables desired and use the `mean` method to compute the mean of the values. Finally, we use `rename` method to change the name of the columns as indicated.

```python
mean_key_vars = endes_2019.groupby(["V024"], as_index=False)[["V201","V613","V715","V511"]].mean(
                ).rename(columns = {"V201":"mean_total_children","V613":"mean_ideal_children","V715":"mean_hb_yr_educ","V511":"mean_first_marriage"})
mean_key_vars
```

#### Answer 8:
We use `pd.melt` function to change the DataFrame format from wide to long. With that, we create a dataframe with the three variables `V024` (the one we selected as the id), `variables` and `values`. Also, we change the name of `V024` to `dpto` using rename function.

```python
reshape_mean_key_vars = pd.melt(mean_key_vars, id_vars = ["V024"], var_name = "variables", value_name = "values").rename(columns = {"V024":"dpto"})
reshape_mean_key_vars
```

#### Answer 10:
We merge `endes_2019` with `reshape_mean_key_vars` using merge function. Inside the proper function, we rename the column `dpto` back to `V024` in `reshape_mean_key_vars` so we can select the column `V024` as the key variable for the merge. Also, we choose to make a left join which means that `endes_2019` is used as the base dataframe. The type of merge we make is an `m:m` so for each row in `endes_2019` we are going to have four (number of unique values in `variables` column in `reshape_mean_key_vars`) columns in `final_result`.

```python
final_result = endes_2019.merge(reshape_mean_key_vars.rename(columns={"dpto":"V024"}), on = ['V024'], how = "left", validate = "m:m")
final_result
```
