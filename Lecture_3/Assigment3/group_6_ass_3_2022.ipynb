Get the min, max, sd, n_obs, n_missing for the following columns **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)**. We want a dataframe with the following columns **Variables, Min, Max, Mean, N_obs, N_missing** and sort by the number of missing rows. **Hint: Use `describe` and `pivot` methods**

```python
#Select the columns
df = endes_2019[["V201","V613", "V715", "V511"]]

#Create a variable with the description of the variables
df1 = df.describe(include = 'all')
#Add the news indexes
df1.loc['n_obs'] = len(df)  
df1.loc['n_missing'] = df.isnull().sum()
```

```python
#Select the required indexes and change the name of the index to "variables"
df2 = df1.loc[["min", "max", "std", "mean", "count", "n_missing"]].reset_index().rename(columns = {'index':'Variables'})
#Pivot the dataframe and rename the columns. Then, we sort the data according to number of missing values. 
df2 = df2.pivot_table(columns= 'Variables').rename(columns=str.capitalize).sort_values(by= 'N_missing').rename(columns = {'Count':'N_obs'})
                                                                                                   
                                                                                                   
                                                                        
df2
```

```python
#Another way to pivot: Transponse
df3= df1.loc[["min", "max", "std", "mean", "count", "n_missing"]].T.rename(columns=str.capitalize).sort_values(by= 'N_missing')
df3
```

7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
# convert categorical variables to numeric variables for "V613"
endes_2019.V613 = pd.Categorical(endes_2019.V613)  
endes_2019['V613_code'] = endes_2019.V613.cat.codes
```

```python
# convert categorical variables to numeric variables for "V715"
endes_2019.V715 = pd.Categorical(endes_2019.V715)
endes_2019['V715_code'] = endes_2019.V715.cat.codes
```

```python
# Create mean_key_vars based on endes_2019 

mean_key_vars = endes_2019.groupby(["V024"])[["V201", "V613_code", "V715_code", "V511"]].mean()
mean_key_vars.rename(columns={"V201": "mean_total_children", 
                              "V613_code": "mean_ideal_children", 
                              "V715_code": "mean_hb_yr_educ",
                              "V511":"mean_first_marriage"}, inplace=True)
mean_key_vars
```

8. Reshape `mean_key_vars` from wide to long. We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

```python
# Convert index to column format 
mean_key_vars.reset_index(drop=False, inplace=True) 
```

```python
mean_key_vars 
```

```python
# Use reshape to convert wide to long using melt  
reshape_mean_key_vars = mean_key_vars.melt(id_vars=["V024"], 
                              var_name="variables", 
                              value_name="values").rename({"V024" : "dpto"}, axis = 1 )
```
