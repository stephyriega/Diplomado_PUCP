8. Reshape `mean_key_vars` from wide to long. We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

```python
## with the group by method we group the year then use a function to get the mean
mean_key_vars = endes_2019.groupby(by=["year","V024"])["V201","V613","V715","V511"].mean()

## change column names 
mean_key_vars = mean_key_vars.rename(columns={"V201":"mean_total_children", "V613":"mean_ideal_children", "V715":"mean_hb_yr_educ", "V511":"mean_first_marriage"})

## the reset index is to have the "dpto" as a new column
mean_key_vars.reset_index(inplace = True)

mean_key_vars = mean_key_vars.rename(columns={"V024":"dpto"})

## select value variables as the four columns indicated
reshape_mean_key_vars = pd.melt(mean_key_vars, id_vars=["dpto"], value_vars=["mean_total_children", "mean_ideal_children", "mean_hb_yr_educ", "mean_first_marriage"], var_name="variables", value_name="values")

reshape_mean_key_vars
```

9. Replicate your answers for questions 7 and 8, but in one line of code. Make it the most simple as possible. **NO HINT**

```python
#Here we are replicating the last 2 questions in one line code. We do here several operations on the dataframe: "endes_2019".
#Using the groupby method, the code groups the dataframe by the columns "year" and "V024" and then calculates the mean for the columns "V201","V613","V715","V511" and assigns the result to the variable mean_key_vars. This results in a new dataframe where the rows are grouped by the "year" and "V024" columns and the columns "V201","V613","V715","V511" have the mean value for each group.
#Then it renames the columns of mean_key_vars dataframe with new names: "V201" to "mean_total_children", "V613" to "mean_ideal_children", "V715" to "mean_hb_yr_educ", "V511" to "mean_first_marriage".
#Then it reset the index using reset_index() method, which removes the multi-level index and makes the index a normal column.
#Then it renames the columns of mean_key_vars dataframe with new names: "V024" to "dpto".
#Then using the melt() method, it reshape the dataframe in long format and it keeps the columns passed in id_vars, and melt the rest of the columns in to two columns, "variable" and "value". These columns are passed in value_vars and var_name, value_name respectively.

reshape_mean_key_vars = endes_2019.groupby(["year","V024"])["V201","V613","V715","V511"].mean().rename(columns={"V201":"mean_total_children", "V613":"mean_ideal_children", "V715":"mean_hb_yr_educ", "V511":"mean_first_marriage"}).reset_index().rename(columns={"V024":"dpto"}).melt(id_vars=["dpto"], value_vars=["mean_total_children", "mean_ideal_children", "mean_hb_yr_educ", "mean_first_marriage"], var_name="variables", value_name="values")

```
